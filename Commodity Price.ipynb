## Commodity Prices

```python
import pandas as pd
```

```python
# read json file format
df = pd.read_json(r'C:\Users\CYPHA\OneDrive\Documents\VisualizeGhana\Price_20240111-155114.json')
df
```

```python
# get commodities
commodities = df.loc['dimension'][0]['commodity_price']['category']['label'].values()
commodities
```

```python
# get months of the years
months = df.loc['dimension'][0]['month']['category']['label'].values()
# get an array or list of months
month = [m for m in months]
month
```

```python
# get the commodity prices
values = df.loc['value'][0]
values
```

```python
# getting the prices for each commodity
# there is a total of 280 month
# and the prices are arranged in order of the commodity
cocoa = values[0:280]
diesel = values[280:560]
gold = values[560:840]
oil = values[840:1120]
petrol = values[1120:1400]
# Hence the length of each values variable should be 280
len(diesel)
```

```python
# Now we must get all the gathered data into a DataFrame
# Create an empty DataFrame
dataFrame = pd.DataFrame(index=commodities, columns=month)
dataFrame
```

```python
# populating each row with their respective data
dataFrame.loc[dataFrame.index[0], month] = cocoa
dataFrame.loc[dataFrame.index[1], month] = diesel
dataFrame.loc[dataFrame.index[2], month] = gold
dataFrame.loc[dataFrame.index[3], month] = oil
dataFrame.loc[dataFrame.index[4], month] = petrol
dataFrame
```

```python
# getting columns for the past five years
year_col = month[216:-4]
# getting columns for each year in the five years 
year_2018 = month[216:228]
year_2019 = month[228:240]
year_2020 = month[240:252]
year_2021 = month[252:264]
year_2022 = month[264:276]
year_2022
```

```python
dataFrame[year_2018]
```

```python
# create a DataFrame for the year 2018
year_2018_avg = dataFrame.loc[dataFrame.index[0:5], year_2018]
# calculate the average price for each commodity and assign it to a new column
year_2018_avg['2018_avg'] = year_2018_avg.mean(axis=1)
year_2018_avg
```

```python
# Repeat same process for all five years
# 2019
year_2019_avg = dataFrame.loc[dataFrame.index[0:5], year_2019]
year_2019_avg['2019_avg'] = year_2019_avg.mean(axis=1)
# 2020
year_2020_avg = dataFrame.loc[dataFrame.index[0:5], year_2020]
year_2020_avg['2020_avg'] = year_2020_avg.mean(axis=1)
# 2021
year_2021_avg = dataFrame.loc[dataFrame.index[0:5], year_2021]
year_2021_avg['2021_avg'] = year_2021_avg.mean(axis=1)
# 2022
year_2022_avg = dataFrame.loc[dataFrame.index[0:5], year_2022]
year_2022_avg['2022_avg'] = year_2022_avg.mean(axis=1)
# year_2022_avg
```

```python
# combine the various DataFrames including the average fo the various years
dataFrame_5y = pd.concat([year_2018_avg,year_2019_avg,year_2020_avg,year_2021_avg,year_2022_avg], axis=1)
dataFrame_5y
```

```python
# get only the average prices in the respective years
df = dataFrame_5y.drop(columns = year_col) # year column contains all years excluding the average of the years
df
```

```python
# For easy visualization the data is transposed
df = df.transpose()
df
```

```python
df = df.rename(index=dict(zip(df.index, ['2018','2019','2020','2021','2022'])))
```
